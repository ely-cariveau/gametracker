```python
from howlongtobeatpy import HowLongToBeat
import requests

# initialize variable for counting games
gameCount = 0
# create empty list for game stats
gameStats = []
# create empty dictionary for all games and their playtime
gameTimePlayed = {}
# create empty dictionary for all games with playtime
gamesPlayed = {}
# create empty dictionary for all games with no playtime
gamesNotPlayed = {}

#define function to get game stats
def getGameStats(gameName):
    
    gameCount = ownedGameSearch['response']['game_count']
    gameStats = ownedGameSearch['response']['games']
    
    for game in gameStats:
        #print(game)
        gameTimePlayed[game['name']] = (game['playtime_forever'])
            
    for game in gameTimePlayed:
        if(gameTimePlayed[game] > 0):
            gamesPlayed[game] = gameTimePlayed[game]
        else:
            gamesNotPlayed[game] = gameTimePlayed[game]
            
    if(gameName in gamesPlayed):
        print(gameName + ' time played: ' + str(gamesPlayed[gameName]//60) + ' hours and ' + str(gamesPlayed[gameName]%60) + ' minutes')
    else:
        print('No playtime data available for '+gameName)
    #print(gameTimePlayed['Counter-Strike'])
```
